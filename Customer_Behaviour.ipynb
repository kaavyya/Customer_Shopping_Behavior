```python
df=df.sort_values(by="Customer ID",ascending=True)
df=df.reset_index(drop = True)
```

```python
df.isnull().sum()
```

```python
df["Purchase Amount (USD)"]=df.groupby("Category")["Purchase Amount (USD)"].transform(lambda x:x.fillna(x.median()))
df["Size"]=df["Size"].fillna(df["Size"].mode()[0])
df["Review Rating"]=df.groupby("Category")["Review Rating"].transform(lambda x:x.fillna(x.median()))
df["Previous Purchases"]=df.groupby("Category")["Previous Purchases"].transform(lambda x:x.fillna(x.median()))
```

```python
df.isnull().sum()
```

```python
df.columns=df.columns.str.replace(" ","_")
df=df.rename(columns={"Purchase_Amount_(USD)" : "Purchase_Amount($)"})
```

```python
df.columns
```

```python
df.head()
```

```python
# To create Column Age Group

df["Age_Group"]=pd.cut(
    df["Age"],
    bins=[17,30,45,60,70],
    labels=["Young","Adult","Middle Aged","Senior"]
                       )
```

```python
df[["Age_Group","Age"]].head()
```

```python
# To create column Loyalty Level

df["Loyalty_Level"]=pd.cut(
    df["Previous_Purchases"],
    bins=[0,10,25,40,50],
    labels=["Bronze","Silver","Gold","Platinum"]
                       )
```

```python
df[["Loyalty_Level","Previous_Purchases"]].head()
```

```python
df.drop("Age_Grouup",axis=1,inplace=True)
```

```python
df.to_csv("Customer_Shopping_Behaviour.csv")
```
